Baris pertama mendefinisikan variabel tinggi dan lebar dengan mengambil dimensi gambar keabuan (grayimage) menggunakan .shape. Variabel ini akan digunakan untuk melakukan iterasi melalui setiap piksel dalam gambar.
Baris kedua membuat array sumGray yang berukuran 256 (satu slot untuk setiap intensitas piksel yang mungkin). Nilai-nilai dalam array ini direpresentasikan sebagai jumlah kemunculan setiap intensitas piksel dalam gambar. Array tersebut diinisialisasi dengan nilai nol menggunakan np.zeros(256). Tipe data array juga diubah menjadi int menggunakan .astype(int) untuk memastikan bahwa nilai-nilai tersebut akan bertipe bilangan bulat.
Dua loop for digunakan untuk melakukan iterasi melalui setiap piksel dalam gambar keabuan. Pada setiap iterasi, nilai intensitas piksel diambil dari gambar dan digunakan sebagai indeks untuk menambahkan satu pada slot yang sesuai dalam array sumGray.
Baris terakhir mencetak array sumGray, yang merupakan histogram gambar keabuan. Histogram ini menunjukkan jumlah piksel untuk setiap nilai intensitas (0 hingga 255).

```python
# Hitung histogram dan menampilkan intensitas
tinggi, lebar = grayimage.shape
sumGray = np.zeros(256).astype(int)
# sumGray = np.zeros(256)
for i in range (tinggi):
    for j in range (lebar):
        sumGray[grayimage[i,j]]+=1
print(sumGray)
```

Baris pertama membuat sebuah list histo yang berisi nilai-nilai dari 0 hingga 255 (256 nilai), yang akan digunakan sebagai sumbu x pada histogram. List ini dibuat menggunakan list comprehension [x for x in range(256)], yang menghasilkan daftar nilai dari 0 hingga 255.
Baris kedua menggunakan fungsi plt.bar() untuk membuat plot histogram. Fungsi ini mengambil dua parameter: histo sebagai sumbu x (nilai keabuan) dan sumGray sebagai sumbu y (frekuensi kemunculan setiap nilai keabuan).
Baris ketiga dan keempat menambahkan label sumbu x dan sumbu y ke plot histogram menggunakan plt.xlabel() dan plt.ylabel(). Label-label tersebut adalah "Keabuan" untuk sumbu x dan "Frequency" untuk sumbu y.

Baris pertama menggunakan fungsi plt.bar() untuk membuat plot histogram yang telah dinormalisasi. Histogram dinormalisasi dengan membagi frekuensi kemunculan setiap nilai keabuan (sumGray) dengan jumlah total piksel dalam gambar (tinggi * lebar). 
Baris kedua dan ketiga menambahkan label sumbu x dan sumbu y ke plot histogram yang telah dinormalisasi. Label-label tersebut adalah "Keabuan" untuk sumbu x dan "Normalized Frequency" untuk sumbu y.

Baris pertama mendefinisikan sebuah kernel konvolusi dalam bentuk matriks dengan menggunakan NumPy array. Kernel konvolusi ini digunakan untuk melakukan operasi konvolusi pada gambar.
Baris ketiga menghitung dimensi kernel dengan kernel.shape, lalu mengatur variabel k_height dan k_width untuk ketinggian dan lebar kernel. Kemudian, gambar keabuan (grayimage) dipad menggunakan np.pad() dengan lebar padding sebesar 2. 
Baris kelima mencetak kernel konvolusi ke konsol untuk melihat nilainya.

```python
# mendefinisikan kernel konvolusi
kernel = np.array([
    [1, 2, -1, 1],
    [0, 2,  1, 2],
    [2, 1,  1, 0],
    [1, 0, -1, 0]
])

# Pad gambar
k_height, k_width = kernel.shape
imagepad = np.pad(grayimage, 2)
print(kernel)
```

Baris pertama membuat sebuah array kosong dengan dimensi yang sesuai untuk menyimpan hasil dari operasi konvolusi. Dimensi array ini dihitung dengan mengurangi tinggi dan lebar gambar asli dengan tinggi dan lebar kernel, dan kemudian ditambah satu karena ukuran gambar akan berkurang setelah operasi konvolusi. 
baris kedua menggunakan np.pad() untuk melakukan padding pada gambar keabuan dengan lebar padding sebesar 2. Ini dilakukan agar ukuran gambar yang dihasilkan setelah operasi konvolusi tetap sama dengan gambar asli. 
Baris ketiga mencetak gambar yang telah dipad ke konsol untuk melihat hasilnya. Ini dilakukan untuk memastikan bahwa proses padding telah berhasil dan gambar telah diperluas sesuai dengan yang diharapkan.

```python
# Konvolusi
konvolusi=np.zeros((tinggi-k_height+1,lebar-k_width+1))
imagepad=np.pad(grayimage,2)
print(imagepad)
```

Dua loop for digunakan untuk melakukan iterasi melalui setiap piksel dalam array konvolusi. Untuk setiap iterasi, sebuah subarray dari gambar yang telah dipad (imagepad) diambil dengan ukuran yang sesuai dengan ukuran kernel (k_height dan k_width). 
Hasil operasi konvolusi disimpan dalam array konvolusi. Untuk setiap iterasi, nilai hasil operasi konvolusi dihitung dan disimpan dalam posisi yang sesuai dalam array konvolusi.
Baris kelima mencetak hasil dari operasi konvolusi ke konsol untuk melihat hasilnya.
Baris terakhir menggunakan plt.imshow() untuk menampilkan gambar hasil konvolusi dalam sebuah plot. Argumen cmap='gray' digunakan untuk menunjukkan bahwa gambar yang ditampilkan adalah gambar keabuan, sehingga menggunakan skala abu-abu dalam plotnya.

```python
for i in range (konvolusi.shape[0]):
    for j in range (konvolusi.shape[1]):
        konvolusi[i][j] = np.sum(imagepad[i:i+k_height, j:j+k_width] * kernel)
print(konvolusi)
plt.imshow(konvolusi, cmap='gray')
```

Variabel rotasi dibuat sebagai matriks kosong dengan ukuran yang sama dengan gambar keabuan (tinggi x lebar).
Sudut rotasi (dalam radian) ditetapkan ke variabel sudut, yang dinyatakan sebagai -40 derajat.
Cosinus dan sinus sudut rotasi dihitung.
Posisi tengah gambar ditentukan dengan tinggi_tengah dan lebar_tengah.
Dilakukan iterasi melalui setiap piksel dalam matriks rotasi.
Setiap piksel diputar dengan menggunakan transformasi geometris:
Koordinat piksel diputar menggunakan rumus transformasi rotasi.
Koordinat hasil rotasi dihitung kembali ke koordinat gambar asli.
Jika hasil rotasi berada dalam batas gambar asli, nilai keabuan piksel tersebut disalin ke matriks rotasi.
Setelah operasi rotasi selesai, gambar hasil rotasi ditampilkan menggunakan plt.imshow() dengan menggunakan skala abu-abu.

TRANSLASI
Variabel translasi dibuat sebagai matriks kosong dengan ukuran yang sama dengan gambar rotasi.
Dilakukan iterasi melalui setiap piksel dalam gambar rotasi.
Setiap piksel dipindahkan sesuai dengan nilai translasi yang ditentukan (x_trans dan y_trans).
Jika hasil translasi berada dalam batas gambar asli, nilai keabuan piksel tersebut disalin ke matriks translasi.
Setelah operasi translasi selesai, gambar hasil translasi ditampilkan menggunakan plt.imshow() dengan menggunakan skala abu-abu.

```python
# Rotasi
rotasi = np.zeros((tinggi, lebar)).astype(int)
sudut = np.deg2rad(-40)
cos = np.cos(sudut)
sin = np.sin(sudut)

tinggi_tengah = int((tinggi-1)/2)
lebar_tengah = int((lebar-1)/2)

for i in range(rotasi.shape[0]):
    for j in range(rotasi.shape[1]):
        X = (i-tinggi_tengah)*cos + (j-lebar_tengah)*sin
        Y = -(i-tinggi_tengah)*sin + (j-lebar_tengah)*cos
        X = int(X + tinggi_tengah)
        Y = int(Y + lebar_tengah)
        if X >= 0 and Y >= 0 and X < rotasi.shape[0] and Y < rotasi.shape[1]:
            rotasi[i][j] = grayimage[X][Y]

plt.title("Hasil Rotasi")
plt.imshow(rotasi, cmap="gray")

# Translasi
x_trans = -163
y_trans = 300

translasi = np.zeros((tinggi, lebar)).astype(int)

for i in range(rotasi.shape[0]):
    for j in range(rotasi.shape[1]):
        if i+y_trans >= 0 and i+y_trans < tinggi and j + x_trans >= 0 and j + x_trans < lebar:
            translasi[i+y_trans][j+x_trans] = rotasi[i][j]

plt.imshow(translasi, cmap="gray")
plt.show()
```

```python
# Pencerminan pada sumbu Y
tinggi, lebar = translasi.shape
pencerminan_y = np.zeros((tinggi, lebar)).astype(int)

for i in range(tinggi):
    for j in range(lebar):
        pencerminan_y[i][j] = translasi[i][lebar - 1 - j]

plt.imshow(pencerminan_y, cmap="gray")
plt.show()
```
